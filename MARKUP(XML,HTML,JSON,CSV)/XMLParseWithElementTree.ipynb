# ElementTree:
<b> DOC: https://docs.python.org/3/library/xml.etree.elementtree.html </b>

```python
from xml.etree import ElementTree
import requests
```

```python
response = requests.get('https://lenta.ru/rss')
xml = response.text
xml
```

```python
#Можем передать напрямую в виде строки.
root = ElementTree.fromstring(xml)
root
```

```python
#Берем xml из файла.
# tree = ElementTree.parse(example.xml)
# root = tree.getroot()
```

```python
print(root.tag)
print(root.attrib)
```

```python
for child in root:
    print(child.tag, child.attrib)
```

```python
root[0][0]
```

```python
root[0][0].text
```

```python
for tag in root.iter('item'):
    print(tag)
```

```python
for tag in root.iter('item'): #перебирает все элементы в поддереве.
    for child in tag:
        print(child.text)
    print('-'*100)
```

```python
for tag in root.iter('item'):
    for link in tag.findall('link'): #перебирает всех детей.
        print(link.text)
```

```python
root = ElementTree.Element('item')
guid = ElementTree.SubElement(root, "guid")
guid.text = "https://lenta.ru/news/2035/01/01/Test/"
author = ElementTree.SubElement(root, "author")
author.text = "Иван Иванов"
title = ElementTree.SubElement(root, "title")
title.text = "Тестовый заголовок для новости"
link = ElementTree.SubElement(root, "link")
link.text = "https://lenta.ru/news/2035/01/01/Test/"
description = ElementTree.SubElement(root, "description")
description.text = "Пошел снег на Новый год"
category = ElementTree.SubElement(root, "category")
category.text = "В мире."
ElementTree.dump(root)
```

<h2> XPath </h2>
Так как XML это дерево, то у нас может быть однозначный путь до одного из тегов.
Другими словами, у нас есть документ в формате XML, но нам не нужен он весь, нам нужна какая-то его часть и с помощью к xpath запросов мы можем к этой части обращаться.

<b> DOC: https://docs.python.org/3/library/xml.etree.elementtree.html#xpath-support </b>

```python
root = ElementTree.fromstring(xml)
root
```

```python
root.findall(".")
```

```python
root.findall("*")
```

```python
root.findall("./channel/item/link")
```

```python
root.findall(".//link")[:10]
```
